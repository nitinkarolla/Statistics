### Poisson Distribution

```r
# Prior Distribution 
#Parameters for gamma prior distribution
k <- 8
rate <- 1

Prior <- rgamma(n=500, shape = k, rate = rate)
```

```r
#Data Observed
X <- rpois(500, lambda = 10)
```

```r
#Posterior Distribution is Gamma distribution (Conjugate prior)
k1 <- k + sum(X)
rate1 <- rate + length(X)

Post <- rgamma(n = 500,shape = k1, rate = rate1)
```

```r
plot(density(Post), col = 'green')
```

### Binomial Distribution

```r
# Prior Distribution 
#Parameters for Beta prior distribution
a <- 1
b <- 1

Prior <- rbeta(n=500, shape1 = a, shape2 = b)
```

```r
#Data Observed
m = 100
X <- rbinom(500, size = m, prob =0.45)
```

```r
#Posterior Distribution is Beta distribution (Conjugate prior)
a1 <- a + sum(X)
b1 <- b + (length(X) * m) - sum(X)

Post <- rbeta(n = 500,shape1 = a1, shape2 = b1)
```

```r
plot(density(Post), col = 'green')
```

### Normal Distribution

#### a) Unknown Mean and Known Precision 

```r
# Normal Distribution
# Parameters for Normal prior distribution
mu <- 3
r <- 0.7

Prior <- rnorm(50, mean = mu, sd =  1/ sqrt(r))
```

```r
#Data Observed
X <- rnorm(50, mean = 7, sd =  2)
```

```r
#Posterior Distribution is Normal distribution (Conjugate prior)
mu1 <- (r*mu +length(X)*r*(sum(X)/length(X)))/(r+length(X)*r)
r1 <- r + length(X)*r

Post <- rnorm(50, mean = mu1, sd =  1/sqrt(r1))
```

```r
plot(density(Post), col = 'green')
```

#### b) Known Mean and Unknown Precision

```r
# Parameters for Gamma prior distribution

alpha <- 3
beta <- 1.5

Prior <- rgamma(50, shape = alpha,rate = beta )
```

```r
#Data Observed
m <- 7
X <- rnorm(50, mean = m, sd =  2)
```

```r
#Posterior Distribution is Gamma distribution (Conjugate prior)
alpha1 <- alpha + length(X)/2

beta1 <- beta + (1/2 * sum(X - m)^2)

Post <- rgamma(50, shape = alpha1 ,rate = beta1)
```

```r
plot(density(Post), col = 'green')
```

### Exponential Distribution

```r
# Parameters for Gamma prior distribution
a <- 2
b <- 0.5

Prior <- rgamma(500,shape =  a, rate = b )
```

```r
#Data Observed
X <- rexp(500, rate = 7)
```

```r
#Posterior Distribution is Gamma distribution (Conjugate prior)
a1 <- a + length(X)
b1 <- b + sum(X)

Post <- rgamma(500,a1, b1)
```

```r
plot(density(Post))
```

```r
# Parameters for Pareto prior distribution
w <- 4
alpha <- 3
```

```r
#Data Observed
X <- runif(50, min = 0, max = 5)
```

```r
#Posterior Distribution is Pareto distribution (Conjugate prior)
alpha.new <- alpha + length(X)
w.new <- max(w, max(X))

Post <- rpareto(50, scale = w.new, shape = alpha.new)
```

```r
plot(density(Post))
```
